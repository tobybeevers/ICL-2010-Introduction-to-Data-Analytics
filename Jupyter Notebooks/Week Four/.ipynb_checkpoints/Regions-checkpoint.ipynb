<h2>Pivot Tables</h2>

A pivot table is a statistics tool that summarizes and reorganizes selected columns and rows of data in a spreadsheet or database table to obtain a desired report. The tool does not actually change the spreadsheet or database itself, it simply pivots or turns the data to view it from different perspectives.

Pivot tables are especially useful with large amounts of data that would be time-consuming to calculate by hand. A few data processing functions a pivot table can perform include identifying sums, averages, ranges or outliers. The table then arranges this information in a simple, meaningful layout that draws attention to key values.

Source: https://www.techtarget.com/searchdatamanagement/definition/pivot-table

Theres a lot of code here. Imagine you had a research question. Of all the years there have been athletics, in which year were the most athletes born?
Use this code to find your for yourself, as we'll as consider what other things you can discover with the code.



```python
## Convert the three columns data relating to birth details and explcitly convert them into datetime datatypes

bios['born_date'] = pd.to_datetime(bios['born_date'])
bios['month_born'] = bios['born_date'].dt.month
bios['year_born'] = bios['born_date'].dt.year

## Group by Year born, then month. We are aggregating the count of names. 
## reset index adds integer numbers to each record as a reference.
## Sort by count of athlete names (the more names the more athletes were born then); force it to be descending

bios.groupby([bios['year_born'],bios['month_born']])['name'].count().reset_index().sort_values('name', ascending=False)

```
