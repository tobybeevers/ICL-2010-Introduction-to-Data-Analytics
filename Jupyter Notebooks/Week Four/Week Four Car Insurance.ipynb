```python
## We’d like to start considering an insurance rating for future customers. 
## On the basis that people in sports cars are at a higher risk, 
## create a new column and give all sports car owners a risk rating of 2 and everybody else a rating of 1.

var1['RISK_RATING'] = var1['VEHICLE_TYPE'].apply(lambda x: 2 if 'sport' in x.lower() else 1)
var1
```

<h4>Explanation:</h4>

- **apply Method:** This applies a function to each element in the VEHICLE_TYPE column.
- **Lambda Function:** The function checks if "sport" (case insensitive) is present in the string. If it is, it assigns a risk rating of 2; otherwise, it assigns a rating of 1.
- **lower() Method:** This ensures the search is case insensitive by converting the string to lowercase.

```python
## Using a lambda function, create a new column that creates a flag system for caution about the drivers.
## If they’ve never had an accident they get a green flag. If they’ve had 2 or less it’s Amber. 
## If they’ve had more than that it’s Red.

## You could even add to this and if they’ve had 5 or more give them a pirate flag.

var1['CAUTION_FLAG'] = var1['PAST_ACCIDENTS'].apply(lambda x: 'green' if x == 0 else 
                                                    'amber' if x <= 2 else 'red' if x < 5 else 'pirate')

var1.to_csv('Car_Insurance_Claim with Caution Flag.csv', index=True)
```
